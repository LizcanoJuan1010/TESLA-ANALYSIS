#Machine Learning Model to predic the price

```python
tesla_values = tesla.history(start="2019-01-01",end="2024-07-05")
```

```python
plt.figure(figsize=(20,6))
plt.plot(tesla_values.index, tesla_values['Close'], lw=2)
plt.xlabel("date")
plt.xlabel("date", fontsize=16)
plt.ylabel("Close price (USD)", fontsize=16)
plt.title("TESLA Closing value", fontsize=16);
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
tesla_values.index = pd.to_datetime(tesla_values.index)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
tesla_values['Close'] = scaler.fit_transform(tesla_values[['Close']])
```

```python
tesla_values.index = pd.to_datetime(tesla_values.index).tz_localize(None)
```

```python
train_data = tesla_values.query('Date <  "2022-06-01"').reset_index(drop = False)
test_data  = tesla_values.query('Date >= "2022-06-01"').reset_index(drop = False)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 6))
ax1.plot(train_data.Date, train_data['Close'], lw=2)
ax1.set_xlabel("date", fontsize=16)
ax1.set_ylabel("Close price (USD)", fontsize=16)
ax1.set_title("Training data", fontsize=16)
ax2.plot(test_data.Date, test_data['Close'], c='orange', lw=2)
ax2.set_xlabel("date", fontsize=16)
ax2.set_ylabel("Close price (USD)", fontsize=16)
ax2.set_title("Test data", fontsize=16);
```

```python
X_train = train_data.Date
y_train = train_data['Close']

X_test = test_data.Date
y_test = test_data['Close']

n_features = 1

train_series = y_train.values.reshape((len(y_train), n_features))
test_series  = y_test.values.reshape((len(y_test), n_features))
```

```python
import keras


from keras.preprocessing.sequence import TimeseriesGenerator


look_back = 20

train_generator = TimeseriesGenerator(train_series, train_series,
                                      length        = look_back,
                                      sampling_rate = 1,
                                      stride        = 1,
                                      batch_size    = 10)

test_generator = TimeseriesGenerator(test_series, test_series,
                                      length        = look_back,
                                      sampling_rate = 1,
                                      stride        = 1,
                                      batch_size    = 10)

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

n_neurons  = 4
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(look_back, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse');

model.fit(train_generator,epochs=300, verbose=0);
```

```python
test_predictions  = model.predict(test_generator)
```

```python
plt.figure(figsize=(20,6))
plt.plot(test_data.Date, test_data['Close'], c='orange',label='true values')
plt.plot(X_test[20:],test_predictions, lw=3, c='r',linestyle = '-', label='predictions')
plt.legend(loc="lower left")
plt.xlabel("date", fontsize=16)
plt.ylabel("Close price (USD) (rescaled)", fontsize=16)
plt.title("TESLA Closing value", fontsize=16);
```

```python
n_days = 20
last_n_days = len(test_data)-n_days

plt.figure(figsize=(15,5))
plt.plot(test_data.Date[last_n_days:], test_data['Close'][last_n_days:], c='orange',marker='o',label='true values')
plt.plot(X_test[last_n_days:],test_predictions[last_n_days-20:],lw=1, c='r', marker='o', label='predictions')
plt.legend(loc="upper left")
plt.xlabel("date", fontsize=16)
plt.ylabel("Close price (USD) (rescaled)", fontsize=16)
plt.title("tesla Closing value", fontsize=16)

# shade in the days for better visualization
date2 = tesla_values.tail(1).index.item()
date1 = date2 - pd.Timedelta(days=n_days+7) # days added for weekends etc.
for i in pd.date_range(date1, date2, periods=n_days+8):
    # shade in 0.6 of each day
    plt.axvspan(i,i+pd.Timedelta(days=0.6), facecolor='lightgrey', alpha=0.2)
plt.show();
```

```python
test_data['naive'] = test_data['Close'].shift(1)
```

```python
plt.figure(figsize=(20,6))
plt.plot(test_data.Date, test_data['Close'], c='orange',label='true values')
plt.plot(X_test[20:],test_predictions, lw=3, c='r',linestyle = '-', label='predictions')
plt.plot(test_data.Date,test_data['naive'], lw=1, c='green',linestyle = '-', label='Naïve forecast')
plt.legend(loc="lower left")
plt.xlabel("date", fontsize=16)
plt.ylabel("Close price (USD) (rescaled)", fontsize=16)
plt.title("Tesla Closing value", fontsize=16);
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
future_dates = pd.date_range(start='2025-01-01', end='2026-12-31', freq='D')
future_series = pd.Series(index=future_dates)

# Limit the loop to the available data in test_predictions
for i in range(min(len(future_dates), len(test_predictions) - look_back + 1)):
  new_row = []  # Initialize as an empty list for numerical values only
  for j in range(look_back):
    new_row.append(test_predictions[i+j][0])  # Extract numerical value from test_predictions
  new_row = np.array(new_row)
  new_row = new_row.reshape((1, look_back, n_features))
  prediction = model.predict(new_row)[0][0]
  future_series[i] = prediction

plt.figure(figsize=(20,6))
plt.plot(test_data.Date, test_data['Close'], c='orange',label='true values')
plt.plot(X_test[20:],test_predictions, lw=3, c='r',linestyle = '-', label='predictions')
plt.plot(future_dates, future_series, lw=1, c='g',linestyle = '--', label='future predictions')
plt.legend(loc="lower left")
plt.xlabel("date", fontsize=16)
plt.ylabel("Close price (USD) (rescaled)", fontsize=16)
plt.title("Tesla Closing value", fontsize=16);
```
